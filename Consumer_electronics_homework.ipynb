```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, classification_report
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('consumer_electronics_sales_data.csv')
data
```

```python
data['ProductCategory'].value_counts()
```

```python
data['ProductBrand'].value_counts()
```

```python
data = pd.get_dummies(data, columns = ['ProductCategory'], dtype = 'int8')
```

```python
data = pd.get_dummies(data, columns = ['ProductBrand'], dtype = 'int8')
```

```python
data
```

```python
data.info()
```

```python
data.describe()
```

```python
y = data['PurchaseIntent']
x = data.drop(['PurchaseIntent'], axis = 1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=1)
```

```python
y_train.shape, y_test.shape
```

```python
sc = StandardScaler()
x_train_sc = pd.DataFrame(sc.fit_transform(x_train), columns = x_train.columns, index = x_train.index)
x_test_sc = pd.DataFrame(sc.transform(x_test), columns = x_test.columns, index = x_test.index)
```

#### Logistic regression

```python
model = LogisticRegression()
model.fit(x_train_sc, y_train)
y_train_pred = model.predict(x_train_sc)
y_test_pred = model.predict(x_test_sc)
print('Class_report_train:')
print(classification_report(y_train, y_train_pred))
print('Class_report_test:')
print(classification_report(y_test, y_test_pred))
```

#### KNN

```python
model_knn = KNeighborsClassifier(n_neighbors=5)
model_knn.fit(x_train_sc, y_train)
y_train_pred = model_knn.predict(x_train_sc)
y_test_pred = model_knn.predict(x_test_sc)
print('Class_report_train:')
print(classification_report(y_train, y_train_pred))
print('Class_report_test:')
print(classification_report(y_test, y_test_pred))
```

### GaussianNB

```python
model_nb = GaussianNB()
model_nb.fit(x_train_sc, y_train)
y_train_pred = model_nb.predict(x_train_sc)
y_test_pred = model_nb.predict(x_test_sc)
print('Class_report_train:')
print(classification_report(y_train, y_train_pred))
print('Class_report_test:')
print(classification_report(y_test, y_test_pred))
```

#### Decision Tree

```python
model_dt = DecisionTreeClassifier(max_depth = 2, min_samples_split=2, min_samples_leaf=1)
model_dt.fit(x_train_sc, y_train)
y_train_pred = model_dt.predict(x_train_sc)
y_test_pred = model_dt.predict(x_test_sc)
print('Class_report_train:')
print(classification_report(y_train, y_train_pred))
print('Class_report_test:')
print(classification_report(y_test, y_test_pred))
```

#### Random Forest

```python
model_rf = RandomForestClassifier(n_estimators=100, max_depth = 2, min_samples_split=2, min_samples_leaf=1)
model_rf.fit(x_train_sc, y_train)
y_train_pred = model_rf.predict(x_train_sc)
y_test_pred = model_rf.predict(x_test_sc)
print('Class_report_train:')
print(classification_report(y_train, y_train_pred))
print('Class_report_test:')
print(classification_report(y_test, y_test_pred))
```
