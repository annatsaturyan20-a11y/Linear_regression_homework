```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

#### Ներմուծենք data-ն

```python
data = pd.read_csv("house_price_regression_dataset.csv")
data
```

#### Տեսնենք, թե ինչ data ենք պարունակում

```python
data.info()
```

#### Համապարփակ ինֆորմացիան նայենք describe-ի միջոցով

```python
data.describe()
```

#### Data-ից երևում է (max) արժեքից, որ Num_Bedrooms, Num_Bathrooms, Neighborhood_Quality-ն կարող ենք սարքել int8 տեսակի, իսկ Lot_Size-ը՝ float16

```python
data['Num_Bedrooms'] = data['Num_Bedrooms'].astype("int8")
data['Num_Bathrooms'] = data['Num_Bathrooms'].astype("int8")
data['Neighborhood_Quality'] = data['Neighborhood_Quality'].astype("int8")
data['Lot_Size'] = data['Lot_Size'].astype("float16")
```

```python
data.info()
```

```python
x = data.drop("House_Price", axis = 1)
y = data['House_Price']
```

#### Կատարենք train_test_split-ը

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.25, random_state=25
)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

#### Հաշվենք հետևյալ գործակիցները․

```python
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f'mse = {mse}\nmae = {mae}\nr2 = {r2}')
```

#### Մոդելի միջին բացարձակ սխալը (MAE) կազմում է 8150, ինչը նշանակում է, որ մոդելը տան գինը միջինում սխալվում է մոտ 8150 միավորով։ R² ցուցանիշը հավասար է 0.9984, ինչը նշանակում է, որ մոդելը բացատրում է գների տատանումների մոտ 99.8%-ը։
