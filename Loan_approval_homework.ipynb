```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, classification_report
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('loan_approval.csv')
data
```

```python
data = data.drop(['name','city'], axis = 1)
data
```

```python
data['loan_approved'] = (data['loan_approved']==False).astype('int8')
```

```python
data
```

```python
data.info()
```

```python
data.describe()
```

```python
y = data['loan_approved']
x = data.drop(['loan_approved'], axis = 1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size = 0.2, random_state = 1)
```

```python
y_train.shape, y_test.shape
```

```python
sc = StandardScaler()
x_train_sc = pd.DataFrame(sc.fit_transform(x_train), columns = x_train.columns, index = x_train.index)
x_test_sc = pd.DataFrame(sc.transform(x_test), columns = x_test.columns, index = x_test.index)
```

```python
model = LogisticRegression()
model.fit(x_train_sc, y_train)
y_train_pred = model.predict(x_train_sc)
y_test_pred = model.predict(x_test_sc)
print(f"accuracy_train: {accuracy_score(y_train, y_train_pred)}, accuracy_test: {accuracy_score(y_test, y_test_pred)}")
print(f"recall_train: {recall_score(y_train, y_train_pred)}, recall_test: {recall_score(y_test, y_test_pred)}")
print(f"precision_train: {precision_score(y_train, y_train_pred)}, precision_test: {accuracy_score(y_test, y_test_pred)}")
print(f"f1_train: {f1_score(y_train, y_train_pred)}, f1_test: {f1_score(y_test, y_test_pred)}")
```

```python
print(classification_report(y_train, y_train_pred))
```

```python
model_knn = KNeighborsClassifier(n_neighbors=3)
model_knn.fit(x_train_sc, y_train)
y_train_pred = model_knn.predict(x_train_sc)
y_test_pred = model_knn.predict(x_test_sc)
print(f"accuracy_train: {accuracy_score(y_train, y_train_pred)}, accuracy_test: {accuracy_score(y_test, y_test_pred)}")
print(f"recall_train: {recall_score(y_train, y_train_pred)}, recall_test: {recall_score(y_test, y_test_pred)}")
print(f"precision_train: {precision_score(y_train, y_train_pred)}, precision_test: {accuracy_score(y_test, y_test_pred)}")
print(f"f1_train: {f1_score(y_train, y_train_pred)}, f1_test: {f1_score(y_test, y_test_pred)}")
```

```python
model_nb= GaussianNB()
model_nb.fit(x_train_sc, y_train)
y_train_pred = model_nb.predict(x_train_sc)
y_test_pred = model_nb.predict(x_test_sc)
print(f"accuracy_train: {accuracy_score(y_train, y_train_pred)}, accuracy_test: {accuracy_score(y_test, y_test_pred)}")
print(f"recall_train: {recall_score(y_train, y_train_pred)}, recall_test: {recall_score(y_test, y_test_pred)}")
print(f"precision_train: {precision_score(y_train, y_train_pred)}, precision_test: {accuracy_score(y_test, y_test_pred)}")
print(f"f1_train: {f1_score(y_train, y_train_pred)}, f1_test: {f1_score(y_test, y_test_pred)}")
```

```python
model_dt= DecisionTreeRegressor(max_depth=3, min_samples_split=2, min_samples_leaf=1, random_state=0)
model_dt.fit(x_train_sc, y_train)
y_train_pred = model_dt.predict(x_train_sc)
y_test_pred = model_dt.predict(x_test_sc)
print(f"accuracy_train: {accuracy_score(y_train, y_train_pred)}, accuracy_test: {accuracy_score(y_test, y_test_pred)}")
print(f"recall_train: {recall_score(y_train, y_train_pred)}, recall_test: {recall_score(y_test, y_test_pred)}")
print(f"precision_train: {precision_score(y_train, y_train_pred)}, precision_test: {accuracy_score(y_test, y_test_pred)}")
print(f"f1_train: {f1_score(y_train, y_train_pred)}, f1_test: {f1_score(y_test, y_test_pred)}")
```

```python
model_rf= RandomForestClassifier(n_estimators=100, max_depth=3, min_samples_split=2, min_samples_leaf=1, random_state=0)
model_rf.fit(x_train_sc, y_train)
y_train_pred = model_rf.predict(x_train_sc)
y_test_pred = model_rf.predict(x_test_sc)
print(f"accuracy_train: {accuracy_score(y_train, y_train_pred)}, accuracy_test: {accuracy_score(y_test, y_test_pred)}")
print(f"recall_train: {recall_score(y_train, y_train_pred)}, recall_test: {recall_score(y_test, y_test_pred)}")
print(f"precision_train: {precision_score(y_train, y_train_pred)}, precision_test: {accuracy_score(y_test, y_test_pred)}")
print(f"f1_train: {f1_score(y_train, y_train_pred)}, f1_test: {f1_score(y_test, y_test_pred)}")
```
