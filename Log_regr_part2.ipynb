#### Logistic regression part 2

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score, classification_report
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('heart.csv')
data
```

```python
data.info()
```

```python
data.describe()
```

```python
data['Sex'] = (data['Sex'] == 'M').astype('int8')
```

```python
data = pd.get_dummies(data, columns=['ChestPainType'], dtype = int)
```

```python
data = pd.get_dummies(data, columns = ['RestingECG'],dtype = int)
```

```python
data = pd.get_dummies(data, columns = ['ExerciseAngina'],dtype = int)
```

```python
data = pd.get_dummies(data, columns = ['ST_Slope'],  dtype = int)
```

```python
data
```

```python
y = data['HeartDisease']
x = data.drop(['HeartDisease'], axis = 1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size= 0.2, random_state= 1)
```

```python
y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
x_train_sc = pd.DataFrame(sc.fit_transform(x_train), columns = x_train.columns, index = x_train.index)
x_test_sc = pd.DataFrame(sc.transform(x_test), columns = x_test.columns, index = x_test.index)
```

#### Logistic Regression

```python
model = LogisticRegression()
model.fit(x_train_sc, y_train)
y_train_pred = model.predict(x_train_sc)
y_test_pred = model.predict(x_test_sc)
```

```python
TP_test = TN_test = FP_test = FN_test = 0

for real, pred in zip(y_test, y_test_pred):

    if real == 1 and pred == 1:
        TP_test += 1
        
    elif real == 0 and pred == 0:
        TN_test += 1
        
    elif real == 0 and pred == 1:
        FP_test += 1
        
    elif real == 1 and pred == 0:
        FN_test += 1

print("TP_test =", TP_test)
print("TN_test =", TN_test)
print("FP_test =", FP_test)
print("FN_test =", FN_test)
```

```python
TP_train = TN_train = FP_train = FN_train = 0

for real, pred in zip(y_train, y_train_pred):

    if real == 1 and pred == 1:
        TP_train += 1
        
    elif real == 0 and pred == 0:
        TN_train += 1
        
    elif real == 0 and pred == 1:
        FP_train += 1
        
    elif real == 1 and pred == 0:
        FN_train += 1

print("TP_train =", TP_train)
print("TN_train =", TN_train)
print("FP_train =", FP_train)
print("FN_train =", FN_train)
```

```python
acc_train = (TP_train + TN_train)/(TP_train + TN_train + FP_train + FN_train)
recall_train = TP_train / (TP_train + FN_train)
precision_train = TP_train / (TP_train + FP_train)
```

```python
f1_train = 2 * recall_train * precision_train / (recall_train + precision_train)
```

```python
print("Accuracy_train: " , acc_train)
print("Recall_train: " , recall_train)
print("Precision_train: " , precision_train)
print("f1_train: " , f1_train)
```

```python
acc_test = (TP_test + TN_test)/(TP_test + TN_test + FP_test + FN_test)
recall_test = TP_test / (TP_test + FN_test)
precision_test = TP_test / (TP_test + FP_test)

f1_test = 2 * recall_test * precision_test / (recall_test + precision_test)

print("Accuracy_test: " , acc_test)
print("Recall_test: " , recall_test)
print("Precision_test: " , precision_test)
print("f1_test: " , f1_test)
```

#### Հիմա report-ով ստուգենք

```python
print(classification_report(y_train, y_train_pred))
print(classification_report(y_test, y_test_pred))
```

```python
model_knn = KNeighborsClassifier(n_neighbors=13)
model_knn.fit(x_train_sc, y_train)
y_train_pred = model_knn.predict(x_train_sc)
y_test_pred = model_knn.predict(x_test_sc)
print(classification_report(y_train, y_train_pred))
print(classification_report(y_test, y_test_pred))
```

```python
model_nb = GaussianNB()
model_nb.fit(x_train_sc, y_train)
y_train_pred = model_nb.predict(x_train_sc)
y_test_pred = model_nb.predict(x_test_sc)
print(classification_report(y_train, y_train_pred))
print(classification_report(y_test, y_test_pred))
```
